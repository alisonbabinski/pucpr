```python
#Dataframe com todas as datas
start = datetime.datetime(year= 2010, month=1, day=1)
end = datetime.datetime(year= 2025, month=12, day=25)
df = pd.DataFrame()
df["data"] = pd.date_range(start, end)
df.head()
```

```python
df_fer = pd.merge(df,df_feriados, how="outer" ,left_on = "data", right_on="dt_fer")
df_fer
```

```python
df_com= pd.merge(df,df_comemorativas, how="outer" ,left_on = "data", right_on="dt_com")
df_com.head(2551)
```

```python
for index,rows in df_fer.iterrows():
  if rows.anterior_feriado == 0.0:
      #print(index, rows.dt_fer, rows.anterior_desc)
      dist = 0
      ferAtual = rows.anterior_desc
  else:
      dist = dist + 1
      df_fer.at[index,"anterior_feriado"] = dist
      df_fer.at[index,"anterior_desc"] = ferAtual

df_fer.head(50)
```

```python
comAtual = "Natal"
#anterior_dist = 7
for index,rows in df_com.iterrows():
  if rows.anterior_dist == 0.0:
      #print(index, rows.dt_fer, rows.anterior_desc)
      dist = 0
      comAtual = rows.anterior_com
  else:
      dist = dist + 1
      df_com.at[index,"anterior_dist"] = dist
      df_com.at[index,"anterior_com"] = comAtual

df_com.head(2551)
```

```python
reversed_fer = df_fer.iloc[::-1]
reversed_fer.head(5)

for index,rows in reversed_fer.iterrows():
  if rows.anterior_feriado == 0.0:
      #print(index, rows.dt_fer, rows.anterior_desc)
      dist = 0
      ferAtual = rows.anterior_desc
  else:
      dist = dist + 1
      reversed_fer.at[index,"posterior_feriado"] = dist
      reversed_fer.at[index,"posterior_desc"] = ferAtual

reversed_fer.head(50)
```

```python
df_fer["feriado"] = False
for index,rows in df_fer.iterrows():
  if rows.dt_fer == rows.data:
      #print(index, rows.dt_fer, rows.anterior_desc)
      df_fer.at[index,"feriado"] = True

df_fer = df_fer.drop(columns=['dt_fer'])
df_fer.head(5)
```

```python
df_com["comemorativo"] = False
for index,rows in df_com.iterrows():
  if rows.dt_com == rows.data:
      #print(index, rows.dt_fer, rows.anterior_desc)
      df_com.at[index,"comemorativo"] = True

df_com = df_com.drop(columns=['dt_com'])
df_com.head(50)
```
